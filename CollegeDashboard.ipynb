```python
# modules
import numpy as np 
import pandas as pd # data import
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact # for interactive plot
%matplotlib inline
```

```python
# import college data
colleges = pd.read_csv("./college_earnings_data.csv")
colleges.head()
```

```python
@interact
def create_earnings_plot(x_variable = colleges.drop(['Mean Earnings (10 Yrs after Entry)','MeanEarningsGenderDiff10','School Name'],
                                                    axis = 1).columns):
    sns.set_style("whitegrid")
    print(x_variable.isnumeric())
    x=x_variable
    print(x.isnumeric())
    if x_variable.isnumeric() == True:
        sns.scatterplot(data = colleges, x = x_variable, y = 'Mean Earnings (10 Yrs after Entry)')
        plt.title(f'Mean Student Earnings by {x_variable}')
    
    elif x_variable.isalpha():
        sns.boxplot(data=colleges,x=x_variable, y = 'Mean Earnings (10 Yrs after Entry)')
        plt.title(f'Mean Student Earnings by College {x_variable}')
        
    
    
        
    
```
